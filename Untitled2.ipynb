```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# visualization settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv("csao_session_dataset.csv")
df.head()
```

```python
df.columns = df.columns.str.lower().str.strip()
```

```python
df['session_timestamp'] = pd.to_datetime(df['session_timestamp'])
```

```python
df['hour'] = df['session_timestamp'].dt.hour
df['day'] = df['session_timestamp'].dt.day_name()
df['month'] = df['session_timestamp'].dt.month
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['user_id'].nunique()
```

```python
df['restaurant_name'].nunique()
```

```python
df['user_city'].unique()
```

```python
df['user_segment'].value_counts()
```

```python
city_counts = df['user_city'].value_counts()

sns.barplot(x=city_counts.values, y=city_counts.index)
plt.title("Sessions by City")
plt.xlabel("Number of Sessions")
plt.ylabel("City")
plt.show()
```

```python
top_restaurants = df['restaurant_name'].value_counts().head(10)

sns.barplot(x=top_restaurants.values, y=top_restaurants.index)
plt.title("Top 10 Restaurants by Sessions")
plt.xlabel("Number of Sessions")
plt.ylabel("Restaurant")
plt.show()
```

```python
sns.countplot(data=df, x='user_segment')
plt.title("User Segment Distribution")
plt.show()
```

```python
cuisine_counts = df['user_preferred_cuisine'].value_counts()

sns.barplot(x=cuisine_counts.values, y=cuisine_counts.index)
plt.title("Preferred Cuisine Distribution")
plt.show()
```

```python
sns.countplot(data=df, x='user_veg_preference')
plt.title("Veg vs Non-Veg Preference")
plt.show()
```

```python
df['user_price_sensitivity'].value_counts()
```

```python
sns.histplot(df['user_price_sensitivity'], bins=20)
plt.title("User Price Sensitivity Distribution")
plt.xlabel("Price Sensitivity")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df['hour'], bins=24)
plt.title("Sessions by Hour of Day")
plt.xlabel("Hour")
plt.show()
```

```python
sns.countplot(data=df, x='day')
plt.title("Sessions by Day of Week")
plt.xticks(rotation=45)
plt.show()
```

```python
df['user_city'].unique()
```

```python
city_counts = df['user_city'].value_counts()

sns.barplot(x=city_counts.values, y=city_counts.index)
plt.title("Food Delivery Sessions by City")
plt.xlabel("Number of Sessions")
plt.ylabel("City")
plt.show()
```

```python
df['user_preferred_cuisine'].unique()
```

```python
cuisine_counts = df['user_preferred_cuisine'].value_counts()

sns.barplot(
    x=cuisine_counts.values,
    y=cuisine_counts.index,
    palette="viridis"
)

plt.title("User Preferred Cuisine Distribution")
plt.xlabel("Number of Sessions")
plt.ylabel("Cuisine")
plt.show()
```

```python
df['user_veg_preference'].unique()
```

```python
df['veg_label'] = df['user_veg_preference'].map({0:'Non-Veg', 1:'Veg'})

sns.countplot(data=df, x='veg_label')

plt.title("Vegetarian vs Non-Vegetarian Preference")
plt.xlabel("Food Preference")
plt.ylabel("Number of Sessions")

plt.show()
```

```python
df['user_segment'].unique()
```

```python
plt.figure(figsize=(8,5))

segment_counts = df['user_segment'].value_counts()

sns.barplot(
    x=segment_counts.index,
    y=segment_counts.values,
    hue=segment_counts.index,
    palette="viridis",
    legend=False
)

plt.title("User Segment Distribution")
plt.xlabel("User Segment")
plt.ylabel("Number of Sessions")

plt.show()
```
